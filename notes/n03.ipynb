```python
# エルボー法で最適なクラスタ数を探索する
max_cluster_num = 50
data_array = dataset_sentence["train"].copy()

sse = []     # Sum of Square Error（クラスタ内誤差平方和）
se = []      # Sum of Error（クラスタ内誤差和）
inertia = [] # scikit-learnにより自動計算するクラスタ内誤差平方和
for i in range(10, max_cluster_num, 5):
    cluster_num = i + 1
    kmeans = KMeans(n_clusters=cluster_num)
    # 各データがどのクラスタに所属するか
    pred = kmeans.fit_predict(data_array)
    inertia.append(kmeans.inertia_)

    # 各データが自身の所属するクラスタ中心からどれだけ離れているか調べる
    transforms = kmeans.transform(data_array)
    distances = np.zeros((data_array.shape[0]))
    for index in range(len(transforms)):
        distances[index] = transforms[index,pred[index]]

    se.append(distances.sum())
    sse.append((distances**2).sum())

    print(f"cluster size {i} done.")
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import argparse
import json
import math
import os
import random
from typing import Dict, List, Tuple


def f01__get_search_space() -> Dict[str, Dict[str, List]]:
  """LightGBM / XGBoost の探索空間（候補値のリスト）を返す"""
  search_space = {
    "LightGBM": {
      "objective": ["binary"],
      "n_estimators": [100, 150, 200, 250, 300],
      "learning_rate": [0.005, 0.01, 0.02, 0.05, 0.1],
      "num_leaves": [2, 4, 8, 16, 32],
      "max_depth": [-1, 4, 6, 8, 10],
      "min_child_samples": [5, 10, 20, 50, 100],
      "subsample": [0.6, 0.8, 1.0],
      "colsample_bytree": [0.6, 0.8, 1.0],
      "subsample_freq": [0, 1, 5, 7],
      "reg_alpha": [0.0, 0.1, 0.5, 1.0, 2.0, 5.0],
      "reg_lambda": [0.0, 0.1, 0.5, 1.0, 2.0, 5.0],
      "min_split_gain": [0.0, 0.1, 0.5],
      "max_bin": [255, 511]
    },
    "XGBoost": {
      "objective": ["binary:logistic"],
      "n_estimators": [100, 150, 200, 250, 300],
      "learning_rate": [0.005, 0.01, 0.02, 0.05, 0.1],
      "max_depth": [3, 4, 6, 8, 10],
      "min_child_weight": [1, 2, 5, 10],
      "subsample": [0.2, 0.4, 0.6, 0.8],
      "colsample_bytree": [0.6, 0.8, 1.0],
      "colsample_bylevel": [0.6, 0.8, 1.0],
      "gamma": [0.0, 0.1, 0.5, 1.0],
      "reg_alpha": [0.0, 0.1, 0.5, 1.0, 2.0, 5.0],
      "reg_lambda": [0.0, 0.1, 0.5, 1.0, 2.0, 5.0],
      "max_delta_step": [0, 1, 3]
    },
  "RFC": {
        # 学習制御
        "n_estimators": [200, 500, 1000, 2000],
        "criterion": ["gini", "entropy", "log_loss"],  # sklearn>=1.2
        "max_depth": [None, 4, 6, 8, 10],
        "min_samples_split": [2, 5, 10, 20],
        "min_samples_leaf": [1, 2, 5, 10],
        "min_weight_fraction_leaf": [0.0],
        # 特徴量サブサンプリング
        "max_features": ["sqrt", "log2", None],  # もしくは [0.5, 0.7, 1.0] などの比率
        "max_leaf_nodes": [None],
        "min_impurity_decrease": [0.0, 1e-4, 1e-3, 1e-2],
        # バギング系
        "bootstrap": [True, False],
        "oob_score": [False, True],  # bootstrap=True の時のみ有効
        "max_samples": [None, 0.5, 0.8, 1.0],    # bootstrap=True の時のみ有効
        # 正則化・重み
        "class_weight": [None, "balanced", "balanced_subsample"],
        "ccp_alpha": [0.0, 1e-4, 1e-3, 1e-2],
        # 実験の安定性
        "random_state": [42]  # 固定 or 必要に応じて外す
      }
    }
  return search_space


def f02__sample_params(rng: random.Random, space: Dict[str, List]) -> Dict:
  """探索空間から 1 サンプル（各パラメータ1値）を無作為抽出"""
  sampled = {}
  for k, v in space.items():
    sampled[k] = rng.choice(v)
  return sampled


def f03__build_experiment_entry(model_name: str, params: Dict) -> Dict:
  """1 実験エントリ（要求仕様の JSON 構造）を作成"""
  return {
    "model_name": "CluCla",
    "hyperparameters": {
      "model_name": model_name,
      "hyperparameter": params
    }
  }


def f04__decide_model_sequence(n_exp: int, models: Tuple[str, ...]) -> List[str]:
  """実験数に対して、モデルを交互に割り当てるシーケンスを生成"""
  seq = []
  for i in range(n_exp):
    seq.append(models[i % len(models)])
  return seq


def f05__build_experiments(n_experiments: int,
                           models: Tuple[str, ...],
                           seed: int) -> List[Dict]:
  """実験一覧を構築（各実験 = 各モデルの単一パラメータサンプル）"""
  rng = random.Random(seed)
  search_space = f01__get_search_space()
  model_seq = f04__decide_model_sequence(n_experiments, models)

  experiments = []
  for m in model_seq:
    params = f02__sample_params(rng, search_space[m])
    experiments.append(f03__build_experiment_entry(m, params))
  return experiments


def f06__to_config(experiments: List[Dict]) -> Dict:
  """最終 JSON 構造に整形"""
  return {
    "handle_imbalance": True,
    "experiments": experiments
  }


def f07__save_json(obj: Dict, path: str) -> None:
  os.makedirs(os.path.dirname(path), exist_ok=True)
  with open(path, "w", encoding="utf-8") as f:
    json.dump(obj, f, ensure_ascii=False, indent=2)


def f08__parse_args() -> argparse.Namespace:
  p = argparse.ArgumentParser(
    description="実験数を指定して LightGBM/XGBoost の単一パラメータセット実験 JSON を生成します（scale_pos_weight は除外）"
  )
  p.add_argument("--n_experiments", type=int, required=True,
                 help="生成する実験数（各実験=単一パラメータセット）")
  p.add_argument("--out", type=str, default="output/04__ML/01__CV_settings.json",
                 help="出力先 JSON パス（既定: output/04__ML/01__CV_settings.json）")
  p.add_argument("--seed", type=int, default=42, help="乱数シード（既定: 42）")
  p.add_argument("--models", type=str, default="LightGBM,XGBoost",
                 help="使用するモデル名をカンマ区切りで（例: LightGBM,XGBoost / LightGBM）")
  return p.parse_args()


def f09__main():
  args = f08__parse_args()
  models = tuple([m.strip() for m in args.models.split(",") if m.strip()])
  if not models:
    raise ValueError("少なくとも 1 つのモデル名を指定してください（例: --models LightGBM ）")

  experiments = f05__build_experiments(
    n_experiments=args.n_experiments,
    models=models,
    seed=args.seed
  )
  config = f06__to_config(experiments)
  f07__save_json(config, args.out)


# ===== ここからノートブックでの実行例（パラメータを編集して実行） =====

# 生成したい実験数
n_experiments = 12

# 使用するモデル（カンマ区切り → タプル化）
models = tuple("LightGBM".split(","))  # 例: ("LightGBM",) だけでもOK

# 乱数シード
seed = 42

# 出力先
out_path = "../output/04__ML/01__config_tmp.json"

# 実行
experiments = f05__build_experiments(
  n_experiments=n_experiments,
  models=models,
  seed=seed,
)
config = f06__to_config(experiments)
f07__save_json(config, out_path)

# 生成結果のプレビュー表示（最初の2件）
print(json.dumps(config, ensure_ascii=False, indent=2)[:2000] + "\n...\n")
print(f"saved: {out_path}")
```
